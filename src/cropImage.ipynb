```python
import numpy as np
import SimpleITK as sitk
import matplotlib.pyplot as plt
sitk.ProcessObject.SetGlobalDefaultNumberOfThreads(4)
```

```python
image = sitk.ReadImage('/Volumes/Untitled/scans/scan.nii.gz')

```

```python
z0 = 1000
z1 = 1450

roi_index = [0, 0, z0]
roi_size  = [image.GetSize()[0], image.GetSize()[1], z1 - z0 + 1]

roi = sitk.RegionOfInterest(
    image,
    size=roi_size,
    index=roi_index
)

sitk.WriteImage(roi, '/Volumes/Untitled/scans/scanCropped.nii.gz')
```
